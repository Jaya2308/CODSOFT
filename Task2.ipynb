```python
from tkinter import *

root = Tk()
root.title("simple calculator")

e = Entry(root, width=45, borderwidth=6)
e.grid(row=0, column=0, columnspan=4, padx=10, pady=10)

e.insert(0, "name")


def click_button(num):
    crnt = e.get()
    e.delete(0, END)
    e.insert(0, str(crnt) + str(num))


def clear_button():
    e.delete(0, END)


def add_button():
    first_num = e.get()
    global f_num
    global math
    math = "add"
    f_num = int(first_num)
    e.delete(0, END)


def equal_button():
    scnd_num = e.get()
    e.delete(0, END)
    if math == "add":
        e.insert(0, f_num + int(scnd_num))
    if math == "sub":
        e.insert(0, f_num - int(scnd_num))
    if math == "mul":
        e.insert(0, f_num * int(scnd_num))
    if math == "div":
        e.insert(0, f_num / int(scnd_num))


def sub_button():
    first_num = e.get()
    global f_num
    global math
    math = "sub"
    f_num = int(first_num)
    e.delete(0, END)


def mul_button():
    first_num = e.get()
    global f_num
    global math
    math = "mul"
    f_num = int(first_num)
    e.delete(0, END)


def div_button():
    first_num = e.get()
    global f_num
    global math
    math = "div"
    f_num = int(first_num)
    e.delete(0, END)


b1 = Button(root, text="1", padx=40, pady=20, command=lambda: click_button(1)).grid(row=3, column=0)
b2 = Button(root, text="2", padx=40, pady=20, command=lambda: click_button(2)).grid(row=3, column=1)
b3 = Button(root, text="3", padx=40, pady=20, command=lambda: click_button(3)).grid(row=3, column=2)
b4 = Button(root, text="4", padx=40, pady=20, command=lambda: click_button(4)).grid(row=2, column=0)
b5 = Button(root, text="5", padx=40, pady=20, command=lambda: click_button(5)).grid(row=2, column=1)
b6 = Button(root, text="6", padx=40, pady=20, command=lambda: click_button(6)).grid(row=2, column=2)
b7 = Button(root, text="7", padx=40, pady=20, command=lambda: click_button(7)).grid(row=1, column=0)
b8 = Button(root, text="8", padx=40, pady=20, command=lambda: click_button(8)).grid(row=1, column=1)
b9 = Button(root, text="9", padx=40, pady=20, command=lambda: click_button(9)).grid(row=1, column=2)
b0 = Button(root, text="0", padx=40, pady=20, command=lambda: click_button(0)).grid(row=4, column=0)

add_button = Button(root, text="+", padx=39, pady=20, command=add_button).grid(row=5, column=0)
equal_button = Button(root, text="=", padx=91, pady=20, command=equal_button).grid(row=5, column=1, columnspan=2)
clear_button = Button(root, text="Clear", padx=79, pady=20, command=clear_button).grid(row=4, column=1, columnspan=2)

sub_button = Button(root, text="-", padx=39, pady=20, command=sub_button).grid(row=5, column=0)
mul_button = Button(root, text="*", padx=39, pady=20, command=mul_button).grid(row=5, column=1, columnspan=2)
div_button = Button(root, text="/", padx=39, pady=20, command=div_button).grid(row=4, column=1, columnspan=2)

root.mainloop()
```
