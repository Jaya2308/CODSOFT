```python
import tkinter
import tkinter.messagebox

window=tkinter.Tk()
window.title("TO-DO-LIST")
tasks=[]
def task_add():
    todo=task_entry.get()
    if todo!="":
        tasks.append(todo)
        update_task_list()
        task_entry.delete(0,tkinter.END)
    else:
        tkinter.messagebox.showwarning(title="Attention",message="click to add task")

def task_remove():
    try:
        index_todo=todo_box.curselection()[0]
        tasks_to_remove=todo_box.get(index_todo)
        tasks.remove(tasks_to_remove)
        update_task_list()
    except IndexError:
        tkinter.messagebox.showwarning(title="Attention",message="delete task")

def edit_task():
    try:
        index_todo = todo_box.curselection()[0]
        task_to_edit = todo_box.get(index_todo)
        task_entry.delete(0, tkinter.END)
        task_entry.insert(tkinter.END, task_to_edit)
        save_edit_button.config(command=lambda: save_task(index_todo))  # Set the command for save_edit_button
        cancel_edit_button.pack(side=tkinter.LEFT)
    except IndexError:
        tkinter.messagebox.showwarning(title="Attention", message="No task selected to edit.")

def save_task(index_todo):
    edited_task = task_entry.get()
    if edited_task!="":
        tasks[index_todo] = edited_task
        update_task_list()
        task_entry.delete(0, tkinter.END)
        cancel_edit_button.pack_forget()
        save_edit_button.config(command=edit_task)  # Restore the command for save_edit_button

def cancel_edit():
    task_entry.delete(0, tkinter.END)
    save_edit_button.config(command=edit_task)  # Restore the command for save_edit_button
    cancel_edit_button.pack_forget()

def update_task_list():
    todo_box.delete(0,tkinter.END)
    for task in tasks:
        todo_box.insert(tkinter.END,task)

list_frame=tkinter.Frame(window)
list_frame.pack()

todo_box=tkinter.Listbox(list_frame,height=20,width=50)
todo_box.pack(side=tkinter.LEFT)

scroller=tkinter.Scrollbar(list_frame)
scroller.pack(side=tkinter.RIGHT,fill=tkinter.Y)

todo_box.config(yscrollcommand=scroller.set)
scroller.config(command=todo_box.yview)

task_entry=tkinter.Entry(window,width=70)
task_entry.pack()

add_task_button=tkinter.Button(window,text="Add task",font=("arial",20,"bold"),background="blue",width=40,command=task_add)
add_task_button.pack()

remove_task_button=tkinter.Button(window,text="delete task",font=("arial",20,"bold"),background="yellow",width=40,command=task_remove)
remove_task_button.pack()

edit_task_button = tkinter.Button(window, text="Edit task", font=("arial", 20, "bold"), background="orange", width=40, command=edit_task)
edit_task_button.pack()

save_edit_button = tkinter.Button(window, text="Save", font=("arial", 20, "bold"), background="green", width=20, command=edit_task)
save_edit_button.pack()

cancel_edit_button = tkinter.Button(window, text="Cancel", font=("arial", 20, "bold"), background="red", width=20, command=cancel_edit)

window.mainloop()
```
